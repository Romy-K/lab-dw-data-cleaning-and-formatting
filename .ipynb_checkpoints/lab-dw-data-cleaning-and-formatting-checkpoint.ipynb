## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv' # enter url to dataframe 
df = pd.read_csv(url) # name dataframe pd and read
df # show first look. Names to be modified: all. 
```

```python
# First, change all letters to lower case and replace spaces with underscores

df.columns = df.columns.str.replace(' ', '_').str.lower()  # make sure to add .columns so as not to change the whole dataframe into just column names!!
df
```

```python
# Replace st with state
df.rename(columns= {'st': 'state'}, inplace=True)
df
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Looking at gender, let's see which values are there 
df.gender.unique()
```

```python
# So I want all 'Femal' and 'female' to join 'F', and all 'Male' to join 'M' 

df.gender = df.gender.replace({'Femal': 'F', 'female': 'F', 'Male': 'M'}) # use a dictionary to replace multiple values at once
df.gender.unique()
```

```python
# Let's see which states we have and do the same thing 
df.state.unique()
```

```python
df.state = df.state.replace({'Cali': 'California', 'AZ': 'Arizona', 'WA': 'Washington'})
df.state.unique()
```

```python
df.education = df.education.replace('Bachelors', 'Bachelor')
df.education.unique()
```

```python
df.customer_lifetime_value = df.customer_lifetime_value.replace('%', '')
df.customer_lifetime_value  # Hmmm this doesn't work. 

df.customer_lifetime_value = df.customer_lifetime_value.str.replace('%', '') # I added str because Google told me to and it works now
df.customer_lifetime_value
```

```python
df.vehicle_class = df.vehicle_class.replace({'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'})
df.vehicle_class.unique()
```

```python
# replace() is my friend
df
```

```python
# Change customer_lifetime_value from str to float (not int! because the NANs are floats) using astype()

df.customer_lifetime_value = df.customer_lifetime_value.astype(float)
df.dtypes
```

```python
# Look at values of number of open complaints

df.number_of_open_complaints.unique()
```

```python
# Change strings with split

# df.number_of_open_complaints = df.number_of_open_complaints.str.split("/")
# df.number_of_open_complaints # error! Are there floats among us? 

```

```python
# We need to create a function and use apply on the column

def get_complaints(value): 
    split = value.split('/')
    if len(split) == 3:
         return float(split[1]) # float because nans are floats, 1 because 1 is the middle part (0 is the first)
    return value # if split != 3 parts, leave as is

df.number_of_open_complaints = df.number_of_open_complaints.astype(str)
df.number_of_open_complaints = df.number_of_open_complaints.apply(get_complaints)  # This gives an error because of the floats in the column, so we need to turn everything into strings first
print(df.number_of_open_complaints)
print(df.number_of_open_complaints.unique())
```

```python
# Now, turn the column into floats

df.number_of_open_complaints = df.number_of_open_complaints.astype(float)
df.dtypes
```

# Bonus: Challenge 3: Analyzing Clean and Formated Data

You have been tasked with analyzing the data to identify potential areas for improving customer retention and profitability. Your goal is to identify customers with a high policy claim amount and a low customer lifetime value.

In the Pandas Lab, we only looked at high policy claim amounts because we couldn't look into low customer lifetime values. If we had tried to work with that column, we wouldn't have been able to because customer lifetime value wasn't clean and in its proper format. So after cleaning and formatting the data, let's get some more interesting insights!

Instructions:

- Review the statistics again for total claim amount and customer lifetime value to gain an understanding of the data.
- To identify potential areas for improving customer retention and profitability, we want to focus on customers with a high policy claim amount and a low customer lifetime value. Consider customers with a high policy claim amount to be those in the top 25% of the total claim amount, and clients with a low customer lifetime value to be those in the bottom 25% of the customer lifetime value. Create a pandas DataFrame object that contains information about customers with a policy claim amount greater than the 75th percentile and a customer lifetime value in the bottom 25th percentile.
- Use DataFrame methods to calculate summary statistics about the high policy claim amount and low customer lifetime value data. To do so, select both columns of the dataframe simultaneously and pass it to the `.describe()` method. This will give you descriptive statistics, such as mean, median, standard deviation, minimum and maximum values for both columns at the same time, allowing you to compare and analyze their characteristics.

```python
df.total_claim_amount.describe()
```

```python
df.customer_lifetime_value.describe()
```

```python
top_claims_start = df.total_claim_amount.quantile(0.75)
top_claims_start

df.top_claims = df.total_claim_amount >= top_claims_start

print(df.top_claims)  # returns boolean, Now I want the Trues to be numbers and the Falses to be NaN 

```

```python
df.top_claims = df.total_claim_amount.where(df.total_claim_amount >= top_claims_start, np.nan)
df.top_claims
```

```python
# Now do the same for the other column, but for lowest 23%

bottom_value_end = df.customer_lifetime_value.quantile(0.25)
df.bottom_value = df.customer_lifetime_value.where(df.customer_lifetime_value <= bottom_value_end, np.nan)
df.bottom_value
```

```python
# Now, join the 2 together using concat

df_expensive_customers = pd.concat([df.top_claims, df.bottom_value], axis=1)
df_expensive_customers

```

```python
# Check for duplicates

df_expensive_customers.duplicated().sum()
```

```python
# Drop duplicates 

df_expensive_customers.drop_duplicates(inplace = True)
df_expensive_customers
```

```python
# Now, we need to deal with the remaining NaNs 

df_expensive_customers.isna().sum()
```

```python
df_expensive_customers.total_claim_amount.describe()  

# I don't think it's wise to just fill in the nans, considering the high standard deviation and range
```
